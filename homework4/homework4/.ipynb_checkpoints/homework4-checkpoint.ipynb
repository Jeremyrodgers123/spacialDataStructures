### Perform regression based on the performance we expect to see. Do tests confirm or disprove our expectations? 

```python
#Dumb KNN
#2 Dimensions
Dumb2D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "Dumb")&
                     (timingData['Dimensions'] == 2)]
Dumb2D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=Dumb2D_U_50K).fit()
print("Dumb R-Squared (2D, Uniform, K=50):", Dumb2D_U_50K_ols.rsquared)

#Buckets
# 2 dimensions
Buckets2D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "Buckets")&
                     (timingData['Dimensions'] == 2)]
Buckets2D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=Buckets2D_U_50K).fit()
print("Buckets R-Squared (2D, Uniform, K=50):", Buckets2D_U_50K_ols.rsquared)

# 3 dimensions
Buckets3D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "Buckets")&
                     (timingData['Dimensions'] == 3)]
Buckets3D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=Buckets3D_U_50K).fit()
print("Buckets R-Squared (3D, Uniform, K=50):", Buckets3D_U_50K_ols.rsquared)

# 10 dimensions
Buckets10D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "Buckets")&
                     (timingData['Dimensions'] == 10)]
Buckets10D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=Buckets10D_U_50K).fit()
print("Buckets R-Squared (10D, Uniform, K=50):", Buckets10D_U_50K_ols.rsquared)


#KD TREE
# 2 dimensions
KdTree2D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "KD Tree")&
                     (timingData['Dimensions'] == 2)]
KdTree2D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=KdTree2D_U_50K).fit()
print("KD Tree R-Squared (2D, Uniform, K=50):", KdTree2D_U_50K_ols.rsquared)

# 3 dimensions
KdTree3D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "KD Tree")&
                     (timingData['Dimensions'] == 3)]
KdTree3D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=KdTree3D_U_50K).fit()
print("KD Tree R-Squared (3D, Uniform, K=50):", KdTree3D_U_50K_ols.rsquared)

# 10 dimensions
KdTree10D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "KD Tree")&
                     (timingData['Dimensions'] == 10)]
KdTree10D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=KdTree10D_U_50K).fit()
print("KD Tree R-Squared (10D, Uniform, K=50):", KdTree10D_U_50K_ols.rsquared)


#Quad Tree
# 2 dimensions
QuadTree2D_U_50K = timingData[(timingData['Distribution'] == 'uniform') & 
                     (timingData['K'] == 50) &
                     (timingData['Type'] == "Quad Tree")&
                     (timingData['Dimensions'] == 2)]
QuadTree2D_U_50K_ols = sm.ols(formula="Time ~ N * np.log(N)", data=QuadTree2D_U_50K).fit()
print("Quad Tree R-Squared (2D, Uniform, K=50):", QuadTree2D_U_50K_ols.rsquared)
```

The rsquared values above show that the run times can be nearly fully explained by N log N. This is because there are N points, and it takes roughly log N amount of time to do the search, which is especially true for the trees.

### What running times do you expect to see based on simple big-O analysis?

#### Dumb KNN
I would expect to see a run time of N Log N with dumb KNN becasue it has to sort the data based on a point and then return the nearest neghbors. Pretty much the best we could do was N Log N with our traditional sorting algorthms.

#### Bucketing
I would expect the bucketing algorithm to have the best lookup time if the size of the bucket is small. Bucketing would have a O(1) lookup time because it is just an array and therefore doesn't have to iterate to find the bucket. Once the algorithm has found the bucket, it just has to iterate through the points in the relevant buckets to find those nearest to K.

#### QuadTree
I would expect the lookup time of the quad tree to be somewhere around k \* log N because at each level of the tree 1/2 of the results can be disguarded if the bucket at that node does not fall within the current radius of the range query. (Note: The radius can change based on our priority Q). I would expect this to perform better than bucketing when the size of the buckets in the bucketing algorithm are large. 

#### KD Tree
I would expect the lookup time of the KD Tree to also be around k \* log N for the same reasons as the Quad Tree. However, because we are always splitting on the median to make our KD Tree, we end up with a perfectly balanced tree where as the Quad tree is likely not balanced. Therefore, we expect the KD Tree to out perform the Quad Tree and Bucketing in situations where the buckets are large or the Quad tree is unbalanced.
